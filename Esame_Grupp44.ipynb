Avendo notato la presenza di nomi di Università che, pur identificando la medesima sede, differiscono per uno o più caratteri, si è deciso di procedere con una normalizzazione degli stessi. Ciò è stato possibile considerando un ulteriore dataset: "school_and_country_table".

Si confrontano i valori delle università con i valori delle università normalizzati.
Se si trova una corrispondenza la funzione restituisce il valore stesso, altrimenti si controllano due eventuali possibilità:
1. La stringa del dataset da normalizzare è contenuta in una di quelle normalizzate: in questo caso la funzione  restituisce quella normalizzata associata.
2. Una stringa tra quelle normalizzate è contenuta nella stringa di partenza: in questo caso la funzione restituisce la stringa normalizzata

In questi ultimi due casi si utilizzano i metodi compile and search.

```python
def normalize(column_element):
    #ciclo for per verificare il match tra le due tabelle
    for university in school_and_country_name['university_name_normalized']:
        if column_element == university:
              return column_element #restituisce l'eventuale elemento trovato

    name_column_element = re.compile(column_element)
    
    for university in school_and_country_name['university_name_normalized']:
        match = name_column_element.search(university)
        #se la stringa name_column_element è compresa in una di school_and_country_name['university_name_normalized']
        #viene restituita quest'ultima.
        if match:
              return university 
        #viceversa se la stringa school_and_country_name['university_name_normalized'] selezionata è compresa in name_column_element
        #viene restituita quest'ultima.
        else:
            name_university = re.compile(university)
            metch_university = name_university.search(column_element)
            if metch_university:
                return university
```

```python
#si applica la funzione precedente a massimoshanghai
massimoshanghai['university_name_normalized'] = massimoshanghai['university_name'].apply(normalize)
massimoshanghai.head()
```

A causa della presenza di range di valori all'interno delle colonne world rank si riscontrano difficoltà relative al calcolo delle differenze tra questi dati. Si rende necessario un trattamento adeguato degli stessi che permetterà successivamente di considerare i valori come elementi numerici

```python
def differenza_max_min(elemento):
    # Match con gli elementi che contengono la dicitura richiesta
    str_times = re.match('(?P<inf>\d+)-(?P<sup>\d+)', elemento['world_rank_times'])
    str_shanghai = re.match('(?P<inf>\d+)-(?P<sup>\d+)', elemento['world_rank_shanghai'])
    
    # se str_times è True, ovvero ho trovato un range, vengono assegnati sup_times e inf_times come estremi del range stesso,
    # altrimenti sup_times e inf_times sono uguali e in particolare corrispondono all'elemento stesso.
    if str_times:
        sup_times = int(str_times.group('sup'))
        inf_times = int(str_times.group('inf'))
    else:
        sup_times = int(elemento['world_rank_times'])
        inf_times = int(elemento['world_rank_times'])
        
    #Per shanghai si procede analogamente a sopra.
    if str_shanghai:
        sup_shanghai = int(str_shanghai.group('sup'))
        inf_shanghai = int(str_shanghai.group('inf'))
    else:
        sup_shanghai = int(elemento['world_rank_shanghai'])
        inf_shanghai = int(elemento['world_rank_shanghai'])
        
    #cwur non contiene range
    cwur_elem = elemento['world_rank']
    
    #Si calcolano massimo e minimo confrontando i valori delle tre tabelle.
    massimo_ = max(sup_times, sup_shanghai, cwur_elem)
    minimo_ = min(sup_times, sup_shanghai, cwur_elem)
    
    # Viene ritornata la differenza fra i valori
    return massimo_ - minimo_
```

```python
times_shanghai_cwur['differenza_max_min'] = times_shanghai_cwur.apply(differenza_max_min, axis=1)
times_shanghai_cwur
```
